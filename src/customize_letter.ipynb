# Using the cover letter utility

This notebook shows the library workflow. For most uses the command line is simpler:

```bash
python utility.py --template example_template.docx --csv example_apps.csv --out ./demo_output
```

Below we do the same thing from Python so you can inspect the intermediate steps.
Run this notebook from the `src/` directory.

## 1. Inspect the applications CSV

Each row is one application. The `slug` column names the output folder; every
other column corresponds to a `${field}` placeholder in the template.

````python
from utility import read_csv_to_dicts, process_csv

apps = read_csv_to_dicts('example_apps.csv')
apps
````

## 2. Generate the letters

`process_csv` fills the template once per row and writes
`cover_letter_<slug>.docx` into a per-application folder under `out_path`.
Re-running never overwrites: an existing `slug` folder gets a `_2`, `_3`, ... suffix.

````python
written = process_csv(
    template='example_template.docx',
    csv_path='example_apps.csv',
    out_path='./demo_output',
)
print(f'Generated {len(written)} letters:', written)
````

## Using your own files

1. Write your cover letter in Word, using `${field_name}` wherever a value should
   change per application (e.g. `${institution_name}`, `${job_title}`).
2. Make a CSV whose header row lists `slug` plus one column per `${field_name}`.
3. Point `template`, `csv_path`, and `out_path` at your files.

The utility writes `.docx` only; export to PDF from Word or LibreOffice when needed.
