# Using the Cover Letter Utility
Let's try it out!

Start with two schools: Goucher College and University of Minnesota. Both are applied positions.

```python
from utility import FileData, PathData, customize_cover_letter, read_csv_to_dicts

# define the necessary data for those two schools in the csv file, read into a dict
APPLICATIONS_NOV15 = read_csv_to_dicts(filepath='dicts_applied_nov15.csv')
# behavioral apps
BEHAVIORAL_APPS = read_csv_to_dicts(filepath='dicts_behavioral.csv')
```

```python
print(APPLICATIONS_NOV15)
```

```python
# instantiate the dataclasses
paths = PathData()
names = FileData()
```

```python
for app in APPLICATIONS_NOV15:
    # make a copy of the original dict and remove the slug to feed into the function
    replace_dict = app.copy()
    replace_dict.pop('slug')

    customize_cover_letter(
        template_name=names.applied_template_name,
        template_path=paths.template_path,
        app_path=paths.app_path,
        slug=app['slug'],
        replace_dict=replace_dict
    )
```

```python
for app in BEHAVIORAL_APPS:
    # make a copy of the original dict and remove the slug to feed into the function
    replace_dict = app.copy()
    replace_dict.pop('slug')

    customize_cover_letter(
        template_name=names.behavioral_template_name,
        template_path=paths.template_path,
        app_path=paths.app_path,
        slug=app['slug'],
        replace_dict=replace_dict
    )
```
